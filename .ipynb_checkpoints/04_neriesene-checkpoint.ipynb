# Jednoduchý súborový manažér

Vytvorte program, ktorý dovolí užívateľovi spravovať zvolený adresár. Najskôr sa na obrazovku vypíše zoznam súborov a adresárov nachádzajúcich sa v aktuálnom pracovanom adresári. Zoznam bude číslovaný a na začiatku musí obsahovať položku `".."` (dve bodky) reprezentujúcu nadradený adresár. Užívateľ má potom možnosť vložiť číslo položky a operáciu, ktorú so zvolenou položkou chce robiť. Zoznam podporovaných operácií:

* `D` - vymaž daný súbor/adresár (pri mazaní si program najskôr vypýta potvrdenie v tvare "Skutocne chces vymazat tuto polozku?")
* `G` - choď do zvoleného adresára (funguje iba pre adresáre, nie pre súbory)
* `S` - zisti veľkosť súboru v bajtoch
* `R` - premenuj daný súbor/adresár (program si vypýta nový názov)
* `C` - vytvor nový adresár
* `N` - vytvor nový prázdny súbor (program si vypýta názov súboru; ošetriť, aby sa nezhodoval s už existujúcim názvom)
* `Q` - ukonči program

Po každej operácii sa musí znova aktualizovať zoznam súborov. V programe ošetrite výnimky (neexistujúce číslo súboru pri vstupe z klávesnice, neexistujúci súbor, vymazanie neprázdneho adresára, atď). Na zistenie obsahu adresára a jeho vypisovanie vytvorte vlastné funkcie. Nedovoľte vymazať adresár, ak obsahuje súbory (použite funkciu `os.rmdir()`, ktorá je bezpečná).
